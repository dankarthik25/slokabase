```python
def get_lyrics(soup):
    simple_text = soup.get_text()  
    # print(simple_text)
     # '\nLYRICS:\n',  '\nLYRICS\n'
    if '\nLYRICS:\n' in simple_text:
        LYRICS = simple_text.split('\nLYRICS:\n')[1]
    elif '\nLYRICS\n' in simple_text:
        LYRICS = simple_text.split('\nLYRICS\n')[1]
    else:
        LYRICS = simple_text.split('LYRICS')[1]
    # 'WORD FOR WORD TRANSLATION: 
    # 'WORD-FOR-WORD\nTRANSLATION:
    # 'SYNONYMS:' 
        
    if '\nWORD\nFOR WORD TRANSLATION:' in LYRICS:         
        LYRICS = LYRICS.split('\nWORD\nFOR WORD TRANSLATION:')[0]
    elif '\nWORD-FOR-WORD\nTRANSLATION:' in LYRICS:
        LYRICS = LYRICS.split('\nWORD-FOR-WORD\nTRANSLATION:')[0]
    elif '\nWORD FOR WORD TRANSLATION:' in LYRICS:
        LYRICS = LYRICS.split('\nWORD FOR WORD TRANSLATION:')[0]
    elif '\nSYNONYMS:' in LYRICS:
        LYRICS = LYRICS.split( '\nSYNONYMS:')[0]
    
    # LYRICS = LYRICS.split('\nWORD\nFOR WORD TRANSLATION:')[0]
    LYRICS = LYRICS.replace('\xa0','').replace('\x92','’').replace('\x91', '‘')
    # LYRICS
        
    import re
    song_list = re.sub(r"\d", "",LYRICS ).replace('()\n','').split('\n\n')

    
    return song_list 
# get_lyrics(simple_text)
```

```python
def get_translation(simple_text):
    # TRASNLATION = simple_text.split('\n\xa0\nTRASNLATION\n\n')[1].split('REMARKS/EXTRA\nINFORMATION:\n')[0]
    # TRASNLATION = simple_text.split('\n\xa0\nTRANSLATION\n1)')[1].split('REMARKS/EXTRA\nINFORMATION:\n')[0]

    if '\n\xa0\nTRANSLATION\n1)' in simple_text:
        TRASNLATION = simple_text.split('\n\xa0\nTRANSLATION\n1)')[1]
    elif 'TRASLATION' in simple_text:
        TRASNLATION = simple_text.split('TRASLATION')[1]
    else:
        TRASNLATION = simple_text.split('\nTRANSLATION')[1]
        
    if '\nREMARKS/EXTRA\nINFORMATION:\n' in TRASNLATION:
        TRASNLATION = TRASNLATION.split('\nREMARKS/EXTRA\nINFORMATION:\n')[0]
    elif '\nRemarks/ Extra Information: ' in TRASNLATION:
        TRASNLATION = TRASNLATION.split('\nRemarks/ Extra Information: ')[0]
    elif 'REMARKS / EXTRA INFORMATION:' in TRASNLATION:
        TRASNLATION = TRASNLATION.split('REMARKS / EXTRA INFORMATION:')[0]
    elif 'REMARKS\n/ EXTRA INFORMATION:' in TRASNLATION:
        TRASNLATION = TRASNLATION.split('REMARKS\n/ EXTRA INFORMATION:')[0]
    elif 'Remarks/ Extra Information:\n' in TRASNLATION:
        TRASNLATION = TRASNLATION.split('Remarks/ Extra Information:\n')[0]
    elif 'REMARKS/EXTRA INFORMATION:' in TRASNLATION:
        TRASNLATION = TRASNLATION.split('REMARKS/EXTRA INFORMATION:')[0]
    elif 'REMARKS/ EXTRA INFORMATION:'  in TRASNLATION:
        TRASNLATION = TRASNLATION.split('REMARKS/ EXTRA INFORMATION:')[0]
    else:
        print('translation end fail')
        print(TRASNLATION)

    import re
    final_trans = re.sub(r"\d\)\n", "",TRASNLATION).split('\n\n')
    for index, index_line in enumerate(final_trans):
        final_trans[index] = index_line.replace('\n',' ').strip()
    if len(final_trans[-1]) == 0 :
        del final_trans[-1]
    # print('\n\n'.join(final_trans))        
    return final_trans
# get_translation(simple_text)
```
