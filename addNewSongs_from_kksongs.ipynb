```python
from slokabase.SqliteModel import get_create_table_query, get_insert_query, get_read_query, get_update_query, get_delete_query
from slokabase.SqliteModel import SqliteModel
```

```python
import os, sqlite3 
# db_name = 'slokabase_10.db'
db_name = 'slokabase.db'
db_path = os.path.join(os.getcwd(),db_name)
```

```python
SongIndex_sql = SqliteModel(db_path,'SongIndex')
mySongs_sql = SqliteModel(db_path,'Songs')
```

```python
# Create Table if Not Created 

SongIndex_sql.create_table(
                    song_idx='INTEGER PRIMARY KEY' , 
                    song_name='VARCHAR(100) NOT NULL UNIQUE', 
                    song_short_name='VARCHAR(100) NOT NULL UNIQUE', 
                    sloka_statues='BOOLEAN NOT NULL CHECK (sloka_statues IN (0, 1))',
                    division='VARCHAR(100)',
                    division_no='INTEGER',
                    total_slokas='INTEGER',
                    groups='VARCHAR(100)',
                    devotion_god='VARCHAR(100)',
                    author='VARCHAR(100)',
                    describtion='TEXT',
                    other_links='TEXT'
                    )


mySongs_sql.create_table(song_idx='INTEGER', 
                    slokas_no='INTEGER NOT NULL', 
                    sloka_hindi= 'TEXT',
                    sloka_eng= 'TEXT',
                    synonyms= 'TEXT',
                    translation= 'TEXT',
                    purpot= 'TEXT',
                    FOREIGN = 'KEY (song_idx) REFERENCES SongIndex(song_idx)',
                    PRIMARY= 'KEY (song_idx, slokas_no)')
```

```python
import csv
prefix_tag = ['sloka_eng', 'synonym', 'translation']
with open('test1.csv', 'r') as data:
    for line in csv.DictReader(data):
        file_path = dict()
        song_data = dict()
        song_name = line['shortname'].strip()
        print(song_name)
        for tag in prefix_tag: 
          file_path[tag] = f"add_song-{tag}/{song_name}-{tag}.txt"
        for tag in prefix_tag:
          with open(file_path[tag],'r') as data:
              song_data[tag]=data.read().split('\n\n')
              
        # add meta data to database
        songIdx_info = dict()
        songIdx_info['song_name'] = line['song_name']
        songIdx_info['song_short_name'] = line['shortname'] # 'SDasaVatStot'
        songIdx_info['sloka_statues'] = 1
        songIdx_info['division_no'] = 0
        songIdx_info['author']= line['author'] 
        songIdx_info['describtion']= f"""Official Name: {line['official_name']}"""
        songIdx_info['other_links '] = f"""lyrics link: {line['lyrics-link']} \n  sysnonym link: {line['sysnonym-link']}"""
        print(songIdx_info)        
        SongIndex_sql.create_entry(**songIdx_info)
        # print(SongIndex_sql.read_entry(*['song_idx'], song_short_name=songIdx_info['song_short_name'])) # >>> 5
        print(SongIndex_sql.read_entry(*['song_idx'], **songIdx_info)[0]['song_idx']) # >>> 5
        
        # add lyrics to database
        
        mysong_info = dict()
        mysong_info ['song_idx'] = SongIndex_sql.read_entry(*['song_idx'], **songIdx_info)[0]['song_idx']
        mysong_info['slokas_no'] = 1
        # mysong_info['sloka_hindi'] =  None
        mysong_info['sloka_eng'] = None
        mysong_info['synonyms'] = None
        mysong_info['translation'] = None
        # mysong_info['purpot']=None
        
        for i in range(0,len(song_data['sloka_eng'])):
            print(i)
            mysong_info['slokas_no']   = i+1
            mysong_info['sloka_eng']   = song_data['sloka_eng'][i].replace('"', '“').replace("'", '’')
            mysong_info['synonyms']    = song_data['synonym'][i].replace('"', '“').replace("'", '’')        
            mysong_info['translation'] = song_data['translation'][i].replace('"', '“').replace("'", '’')
            mySongs_sql.create_entry(**mysong_info)
            # mySongs_sql.update_entry(**mysong_info)
        

        
        # break  
```
